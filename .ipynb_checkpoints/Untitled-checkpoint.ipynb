```python
with open("pacienti.txt") as f:
    content = f.readlines()
#content = [x.strip() for x in content]
patients = [ {"date": y[0].strip(), "type": y[1].strip() } for y in [x.strip().split(",") for x in content] ]
```

```python
def creteDict():
    return {
        "26-02-2020": 0,
        "27-02-2020": 0,
        "28-02-2020": 0,
        "29-02-2020": 0,
        "01-03-2020": 0,
        "02-03-2020": 0,
        "03-03-2020": 0,
        "04-03-2020": 0,
        "05-03-2020": 0,
        "06-03-2020": 0,
        "07-03-2020": 0,
        "08-03-2020": 0,
        "09-03-2020": 0,
        "10-03-2020": 0,
        "11-03-2020": 0,
        "12-03-2020": 0,
        "13-03-2020": 0,
        "14-03-2020": 0,
    }
```

```python
patientsPerDay = creteDict()
patientsPerDayExternal = creteDict()
patientsPerDayInternal = creteDict()
for patient in patients:
    patientsPerDay[patient["date"]] += 1
        
    if patient["type"] == "E":
        patientsPerDayExternal[patient["date"]] += 1
    else:
        patientsPerDayInternal[patient["date"]] += 1
        
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 10))
ax1.bar(patientsPerDayInternal.keys(), patientsPerDayInternal.values())
ax2.bar(patientsPerDayExternal.keys(), patientsPerDayExternal.values())
ax3.bar(patientsPerDay.keys(), patientsPerDay.values())

plt.show()
```

```python
def cumulate(d):
    res = creteDict()
    prev = 0
    for k in d:
        res[k] = prev + d[k]
        prev = res[k]
    return res
        
cumulative = cumulate(patientsPerDay)

lists = cumulative.items() # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.plot(x, y, linestyle='--', marker='o')
#plt.plot(cumulative.keys(), cumulative.values())
```

```python
from scipy.optimize import curve_fit
import datetime
from matplotlib.dates import date2num
import numpy as np

def func(x, a, b, c):
    return a * np.exp(b * x) + c

l = len(patientsPerDayInternal.keys())
x = np.linspace(1, l, l)
y = list(patientsPerDayInternal.values())

popt, pcov = curve_fit(func, x, y)

plt.figure()
plt.plot(x, y, 'ko', label="Original Noised Data")
plt.plot(x, func(x, *popt), 'r-', label="Fitted Curve")
plt.legend()
plt.show()
```
